```python
from transformers import pipeline
```

```python
generator = pipeline("text2text-generation", model="google/flan-t5-base")
```

```python
def generate_resume_prompt(profile, job_desc):
    return f"""Write 5 resume bullet points for this job description using the candidate's profile.

Candidate Profile:
{profile}

Job Description:
{job_desc}

Bullet points:"""

def generate_cover_letter_prompt(profile, job_desc):
    return f"""hellooo 333 4 4445.

Candidate Profile:
{profile}

Job Description:
{job_desc}

Cover letter:"""
```

```python
profile = """
Kartik Marjara
B.E. in CSE (AI & ML) at Chandigarh University
Internship at XYZ in AI/ML, Graphic Lead at GFG CU Chapter
Skills: Python, NLP, Transformers, Streamlit, GitHub
"""

job_desc = """
We are hiring an AI/ML Engineer with NLP experience and skills in LLMs, Python, and deployment. Strong communication skills are a bonus.
"""
```

```python
resume_prompt = generate_resume_prompt(profile, job_desc)
cover_prompt = generate_cover_letter_prompt(profile, job_desc)
```

```python
resume_output = generator(resume_prompt, max_new_tokens=300)[0]["generated_text"]
cover_output = generator(cover_prompt, max_new_tokens=300)[0]["generated_text"]
```

```python
print("🎯 Resume Bullet Points:\n")
print(resume_output.strip())

print("\n📄 Cover Letter:\n")
print(cover_output.strip())
```
